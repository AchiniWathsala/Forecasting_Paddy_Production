```python
def create_date(row):
    if row['Season'].lower() == "yala":
        return pd.Timestamp(year = row['Year'], month=4, day=1)
    elif row['Season'].lower() == "maha":
        return pd.Timestamp(year = row['Year']-1, month=9, day=1)
    
data['Date'] = data.apply(create_date, axis=1)

# sort for time series
data = data.sort_values('Date')

# set as index (important for forecasting)
#data = data.set_index('Date')

data.head()
```
