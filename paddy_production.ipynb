```python
# Change the data type of the variables
data = data.astype({
    # "Year" : ,
    # "Season" : ,
    "Sown_Extent_Major" : "int", 
    "Sown_Extent_Minor" : "int",
    "Sown_Extent_Rainfall" : "int",
    "Sown_Extent_Total" : "int",
    "Harvested_Extent_Major" : "int",
    "Harvested_Extent_Minor" : "int",
    "Harvested_Extent_Rainfall" : "int",
    "Harvested_Extent_Total" : "int",
    "Avg_Yield_Major" : "int",
    "Avg_Yield_Minor" : "int",
    "Avg_Yield_Rainfall" : "int",
    "Avg_Yield_Total" : "int",                  
    "Production" : "float"
})
```

```python
data.info()
```

```python
def create_date(row):
    if row['Season'].lower() == "yala":
        return pd.Timestamp(year = row['Year'], month=4, day=1)
    elif row['Season'].lower() == "maha":
        return pd.Timestamp(year = row['Year']-1, month=9, day=1)
    
data['Date'] = data.apply(create_date, axis=1)

# sort for time series
data = data.sort_values('Date')

# set as index (important for forecasting)
#data = data.set_index('Date')

data.head()
```
